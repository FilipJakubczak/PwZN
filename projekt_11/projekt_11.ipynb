```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist("facebook_combined.txt", create_using=nx.Graph(), nodetype=int)
```

```python
# Rozkład stopni
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()

plot_degree_dist(nx.gnp_random_graph(100, 0.5, directed=False))
```

```python
# Składowe spójne
conn_comp = [len(c) for c in sorted(nx.connected_components(G), key=len, reverse=True)]
conn_comp
# Jak widać sieć posiada jedynie jedną składową spójną, czyli mamy graf połączony
```

```python
# Najkrótsza średnia ścieżka
nx.average_shortest_path_length(G)
```

```python
# Rysowanie sieci
nx.draw(G, with_labels = False)
```
